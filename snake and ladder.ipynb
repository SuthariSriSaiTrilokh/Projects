```python
# SNAKE AND LADDER
import random as rd

class Board:
    def __init__(self) -> None:
        self.board_size = 100
        self.snakes = {17:6, 54:34, 62:19, 64:60, 87:36, 93:73, 95:75, 98:79}
        self.ladders = {1:38, 4:14, 9:31, 21:42, 28:84, 51:67, 72:91, 80:99}
    
    def new_position(self, player, dice_number):
        player.position += dice_number
        if player.position > self.board_size:  # Prevent exceeding the board
            player.position -= dice_number
        elif player.position in self.snakes:
            player.position = self.snakes[player.position]
        elif player.position in self.ladders:
            player.position = self.ladders[player.position]

class Player:
    cnt = 0
    def __init__(self, name):
        self.name = name
        self.position = 0
        Player.cnt += 1

# Game setup
b = Board()  # Create a board
p1 = Player('Player1')
p2 = Player('Player2')
p3 = Player('Player3')
list_of_players = [p1, p2, p3]
win = False
player_index = 0

# Game loop
while not win:
    input('Press enter to roll dice: ')
    dice_number = rd.randint(1, 6)  # Roll the dice
    b.new_position(list_of_players[player_index], dice_number)
    print(f"{list_of_players[player_index].name} ---- {list_of_players[player_index].position} (Dice: {dice_number})")
    
    if list_of_players[player_index].position == b.board_size:
        print(f'Hurray! {list_of_players[player_index].name} won the game!')
        break
    
    if dice_number != 6:  # Only switch turns if dice roll is not 6
        player_index = (player_index + 1) % Player.cnt

print('Game over')
```
